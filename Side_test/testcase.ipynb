```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
plt.figure(figsize=(12,8))
X = np.random.normal(0,1,100)
plt.subplot(2,2,1)
plt.hist(X)
plt.subplot(2,2,2)
stats.probplot(X,dist="norm",plot=plt)
plt.show()
```
